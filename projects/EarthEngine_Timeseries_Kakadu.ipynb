## Extract the fractional Cover data from the timeseries for each site

```python
# Dates from the project sampling spreadsheet
# Given as Excel dates
excelDateList = [34681,34682,34683,34684,34685,34686,34687,34702,34703,34704,34705,34706,34707,34708,34715,34716,34717,34718,34719,34720,34721,34722,34724,34729,34730,34731,34732,34733,34734,34735,34736,34737,34738,34744,34745,34746,34747,34748,34749,34750,34751,34752,34753,34757,34758,34759,34761,34762,34763,34764,34765,34771,34772,34773,34774,34775,34776,34777,34778,34779,34780,34790,34808,34809,34810,34906,34907,34912,34913,34948,34949,34950,35080,35082,36251,36579,36580,36585,36617,36628,36697,36982,37165,38042,38078,38411,38434,38443,38854,39881,39904,41380,41383,41387,41553,41554,41555,41556,41557,41674,41694,41695,41696,41697,41698,41700,41701,41702,41703,41704]
# Convert to datetime
dateList = [datetime.datetime(1899, 12, 31)+ relativedelta(days=sampleDate) for sampleDate in excelDateList]

print "From " + dateList[0].strftime('%Y-%B') + " to " + dateList[-1].strftime('%Y-%B')




# Connect to the field site fusion table: http://goo.gl/ui08FG
#fieldSites = ee.FeatureCollection('ft:14lKbca2-KCVM-46o2i0QJThZOcn-zrVeLg5hX98z')
fieldSites = ee.FeatureCollection('ft:1HZ4TLzpk6Tu78_g879HBMJfHJx0ElrDNyBhvm4mH')

# Initialise the output array
fieldArray = []

# Build the combined reducer
reducerMean = ee.Reducer.mean().forEach(['Bare_Mean','Green_Mean','Dead_Mean'])
reducerSdev = ee.Reducer.stdDev().forEach(['Bare_StDev','Green_StDev','Dead_StDev'])
reducerCombined = reducerMean.combine(reducerSdev, None, True)

# Progress Bar
pBar = progressBar(len(dateList))
counter=0
pBar.animate(counter)





for midDate in dateList:
    # Unmix the image for the selected season
    collection = 'LANDSAT/LT5_L1T_TOA' if midDate < datetime.datetime(2000,1,1)  else 'LANDSAT/LE7_L1T_TOA'
    coverFractions = applyRSCtransforms(nominalMedian(collection,midDate)) \
    .unmix([end_bare,end_gren,end_dead]).select(["band_0","band_1","band_2"])
    # Update the progress bar
    counter+=1
    pBar.animate(counter)
    # Find the mean and standard deviation for all the sites
    summarySites = coverFractions.reduceRegions(fieldSites, reducerCombined, 30).getInfo()['features']
    # Extract the relevant data from the summary sites into an array
    for site in summarySites:
        siteData =  site['properties'] # The feature properties
        excelDate = (midDate - datetime.datetime(1899, 12, 31)).days
        fieldArray.append([excelDate,siteData['OBJECTID'],siteData['Bare_Mean'],\
                           siteData['Green_Mean'],siteData['Dead_Mean'],\
                           siteData['Bare_StDev'],siteData['Green_StDev'],siteData['Dead_StDev'],\
                           siteData['name']])
```

## Make some plots

### Individual

```python
# Convert the data to a numarray
siteFCData = fieldArray[:,0:8].astype(numpy.float32)
siteFCNames = fieldArray[:,8]


# Basic Timeseries Plot
def fcTimePlot(siteFID,tsDataArray,siteFCNames):
    # Select the site of interest
    siteData = tsDataArray[tsDataArray[:,1]==siteFID]
    siteName = siteFCNames[tsDataArray[:,1]==siteFID][0]
    # Recreate the list of dates
    dateList = [datetime.datetime(1899, 12, 31) + relativedelta(days = x) for x in siteData[:,0]]
    # Build the plot
    fig, ax = matplotlib.pyplot.subplots()
    matplotlib.pyplot.tight_layout()
    matplotlib.rcParams.update({'font.size': 14})
    fig.set_size_inches(10,5)
    title("Site %s" % siteName) 
    ylabel('Cover amount')
    xlabel('Year')
    # Add the cover plots
    ax.plot(dateList, siteData[:,2],'red', label="Bare")
    ax.fill_between(dateList, siteData[:,2]+siteData[:,5], siteData[:,2]-siteData[:,5], facecolor='red', alpha=0.2)
    ax.plot(dateList, siteData[:,3],'green', label="Green")
    ax.fill_between(dateList, siteData[:,3]+siteData[:,6], siteData[:,3]-siteData[:,6], facecolor='green', alpha=0.2)
    ax.plot(dateList, siteData[:,4],'blue',label="non-Green")
    ax.fill_between(dateList, siteData[:,4]+siteData[:,7], siteData[:,4]-siteData[:,7], facecolor='blue', alpha=0.2)
    # Make it look nice
    ax.grid(b=True, which='major', color='yellow', linestyle='--')
    ax.legend(loc='upper right', shadow=True)
    ax.set_ylim([0.0,1.0])
    
  
    return ax


for x in numpy.unique(siteFCData[:,1]):
    fcTimePlot(x,siteFCData,siteFCNames)
    matplotlib.pyplot.savefig("/home/ipython/understanding-australias-changing-ecosystems/assets/%d.png" % x)
    


#plotList = [fcTimePlot(x,fieldArray) for x in numpy.unique(fieldArray[:,1])]




```
